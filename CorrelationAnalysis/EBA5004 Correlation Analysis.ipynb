```python
# create hold-out sample
df_seen = df_all_grouped_imputed.sample(frac=0.80, random_state=123)
df_unseen = df_all_grouped_imputed.drop(df_seen.index)
df_seen.reset_index(inplace=True, drop=True)
df_unseen.reset_index(inplace=True, drop=True)
# df_seen_filtered = df_seen[df_seen['stock']=='TSLA']
# df_unseen_filtered = df_unseen[df_unseen['stock']=='TSLA']
```

```python
# feature list for model comparison
# this list is based on the results after the evaluation in order to speed up the model development
dropped_features = ['avg_sentiment_score_predicted',
                     'price_change_same_day'
                    ]

# this list is the masked features in order to compare the performance with and without certain features
without_sentiment = ['avg_sentiment_score',
                     'avg_sentiment_score_combined',
                     'avg_pre_day_volume_zscore'
                    ]

true_sentiment_only = ['avg_sentiment_score_combined',
                       'avg_pre_day_volume_zscore'
                      ]

combined_sentiment = ['avg_sentiment_score',
                      'avg_pre_day_volume_zscore'
                      ]

with_volume = ['avg_sentiment_score',
              ]

masked_feature_list_of_list = [without_sentiment, true_sentiment_only, combined_sentiment, with_volume]
```

```python
# common model setup for batch processing
target='price_change_pre_day'
session_id=123
use_gpu=True
normalize=True
normalize_method='zscore'
fold=5
fix_imbalance=True
fix_imbalance_method=None # default method is SMOTE
categorical_features=['stock', 'within_trading_time',
                      'created_month_local',
                      'created_day_local',
                      'created_weekday_local',
                      'created_hour_local']
imputation_type='iterative'
iterative_imputation_iters=3
silent=True
model_type = 'et'
```

```python
df_test_seen = df_all_grouped_imputed.sample(frac=0.004, random_state=123)
df_test_unseen = df_all_grouped_imputed.sample(frac=0.001, random_state=456)
```

```python
data = df_test_seen
data_unseen = df_test_unseen
%%time
# dataframe to store all results
df_result_all = pd.DataFrame()
# loop with different feature set
for masked_feature_list in masked_feature_list_of_list:
    ignore_features_list = dropped_features + masked_feature_list
    print(data.columns)
    # setup model
    s = setup(data=data,
        target=target,
        session_id=session_id,
        use_gpu=use_gpu,
        normalize=normalize,
        normalize_method=normalize_method,
        fold=fold,
        fix_imbalance=fix_imbalance,
        fix_imbalance_method=fix_imbalance_method,
        categorical_features=categorical_features,
        ignore_features=ignore_features_list,
#         imputation_type=imputation_type,
#         iterative_imputation_iters=iterative_imputation_iters,
        silent=silent
             )
    model = create_model(model_type)
    # finalize
    final_model = finalize_model(model)
    # predict
    unseen_predictions = predict_model(final_model, data=data_unseen)
    df_result = pull()
    df_result_all = pd.concat([df_result_all, df_result], axis=0)
```

# Batch evaluation between features for individual stock

```python
stock_list = df_seen['stock'].unique()
data=df_seen
data_unseen=df_unseen
```

```python
%%time
# dataframe to store all results
df_result_all_stock = pd.DataFrame()
# loop with different feature set
for stock in stock_list:
    for masked_feature_list in masked_feature_list_of_list:
        ignore_features_list = dropped_features + masked_feature_list
        print(data.columns)
        # setup model
        s = setup(data=data[data['stock']==stock],
            target=target,
            session_id=session_id,
            use_gpu=use_gpu,
            normalize=normalize,
            normalize_method=normalize_method,
            fold=fold,
            fix_imbalance=fix_imbalance,
            fix_imbalance_method=fix_imbalance_method,
            categorical_features=categorical_features,
            ignore_features=ignore_features_list,
#             imputation_type=imputation_type,
#             iterative_imputation_iters=iterative_imputation_iters,
            silent=silent
                 )
        model = create_model(model_type)
        # finalize
        final_model = finalize_model(model)
        # predict
        unseen_predictions = predict_model(final_model, data=data_unseen[data_unseen['stock']==stock])
        df_result = pull()
        df_result['stock'] = stock
        df_result_all_stock = pd.concat([df_result_all_stock, df_result], axis=0)
```
